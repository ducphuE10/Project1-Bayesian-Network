# Tutorial

# 1. Node

Trong project này em implement lại các node cơ bản 

Có các loại node sau đây
*   **Label node**: Label_Node(id_name : str, states : list)
*   **Boolean node**: Boolean_Node(id_name : str, states = [True, False])
  *   **OR node**: OR_Node(id_name : str)
  *   **XOR node**: XOR_Node(id_name : str)
  *   **AND node**: AND_Node(id_name : str)
  *   **MtoN node**: MtoN_Node(id_name : str, M: int)
*   **Ranked node**: Ranked_Node(id_name : str, states : list, mean : float, std : float)
  * **Weighted_Node**: Weighted_Node(id_name : str, states: list, weighted_mean : dict, mean = None, std : str)

```python
node1 = Boolean_Node('node1')
node2 = Label_Node('node2',['football', 'tennis','volley ball'])
node3 = Ranked_Node('node3',['very low','low','medium','high','very high'], mean = 0.5, std = 0.5)
node4 = Boolean_Node('node4')
node5 = Weighted_Node('node5', ['very low','low','medium'], weighted_mean = {node1: 0.5, node2:0.5}, std = 1)
```

# 2. Graph

```python
G = Graph()
```

thêm các nodes vào Graph: 

G.add_nodes(*danh sách các nodes)

```python
G.add_nodes(node1,node2,node3,node4)
```

thêm cạnh giữa các nodes vào Graph:

G.add_edges(*danh sách các cạnh)

```python
G.add_edges((node1,node2),(node3,node2),(node2,node4))
```

visualize Graph

```python
G.visualize()
```

```python
G.remove_node(node4)
```

```python
G.visualize()
```

```python
G.visualize()
```

# 3. Network

truyền một Graph vào network để khởi tạo

```python
network = Network(graph = G)
```

Graph truyền vào network phải là đồ thị có hướng KHÔNG CÓ CHU TRÌNH

```python
node_1 = Boolean_Node('node1')
node_2 = Boolean_Node('node2')
node_3 = Label_Node('node3',states = ['a','b','c'])
G_ = Graph()
G_.add_nodes(node_1,node_2,node_3)
G_.add_edges((node_1,node_2),(node_2,node_3),(node_3,node_1))
G_.visualize()
network = Network(graph = G_)
```

tạo NPT cho các nodes thông qua hàm 
help_generate_NPT(list_nodes, path_to_forder)

```python
network.help_generate_NPT([node1,node2],path_to_folder='data/test')
```

NPT của một node được lưu theo file CSV

Điền thủ công các giá trị vào file (function node không cần)

Sử dụng hàm **.set_NPT_from_csv(node: Node, path)** để set NPT cho một node

Sử dụng hàm **.set_NPT_func(func_node: Node)** để set NPT cho một function node

Sau khi gán NPT cho tất cả các node trong mạng:

 * sử dụng **.initial_states()** để xem trạng thái ban đầu của mạng
 * sử dụng **.conditional_probability(observation: dict, conditional: dict)** để tính xác suất điều kiện
 * sử dụng **.marginal_probability(observation: dict)** để tính xác suất biên
 * sử dụng **.set_evidence(self, evidence: dict)** để xem trạng thái của mạng khi set evidence

## 4.4. Obtaining Username and Password

Để đánh cắp thông tin (tài khoản & mật khẩu) của người dùng thì có thể từ:
* **From system**: từ hệ thống (từ bản rõ hoặc decrypt bản mã)
* **Credential Stuffing**:  thông tin xác thực thu được từ một vụ vi phạm dữ liệu trên một dịch vụ, được sử dụng để cố gắng đăng nhập vào một dịch vụ không liên quan khác.
* **Guessing Attack**: Đoán (ví dụ một số mật khẩu phổ biến..)
* **Brute Force Attack**: cố gắng đoán mật khẩu mà không có ngữ cảnh hoặc manh mối, sử dụng các ký tự một cách ngẫu nhiên đôi khi kết hợp với những đề xuất mật khẩu phổ biến
* **From User**: từ người dùng
    * **Social Engineering**:liên quan đến việc thao tác tâm lý của con người 
    * **Key logger**: theo dõi thao tác phím
    * **Cross-site Request**: dựa vào mượn quyền trái phép
    * **Coersive Force**: Cưỡng chế
    * **Stolen Information**: Trộm thông tin

```python
Encrypted_File = Boolean_Node('Encrypted_File')
File_Decrypted = Boolean_Node('File_Decrypted')
PlainText_File = Boolean_Node('PlainText_File')

Credential_Stuffing = Boolean_Node('Credential_Stuffing')
Brute_Force = Boolean_Node('Brute_Force')
Stolen_Infor = Boolean_Node('Stolen_Infor')
Coersive_Force = Boolean_Node('Coersive_Force')
Cross_Site = Boolean_Node('Cross_Site')
Key_Logger = Boolean_Node('Key_Logger')
Social = Boolean_Node('Social')

Obtaining = OR_Node('Obtaining')
From_User = OR_Node('From_User')
From_System = XOR_Node('From_System')
Authen_Vio = OR_Node('Authen_Vio')
```

```python
G = Graph()
G.add_nodes(Encrypted_File,File_Decrypted,PlainText_File,From_System,Credential_Stuffing,Brute_Force,From_User,Stolen_Infor,Coersive_Force,Cross_Site,Key_Logger,Social,Obtaining,Authen_Vio)
```

```python
G.add_edges((Encrypted_File,File_Decrypted),
            (PlainText_File,From_System),
            (File_Decrypted,From_System),
            (From_System,Obtaining),
            (Credential_Stuffing,Obtaining),

            (Brute_Force,Obtaining),
            (From_User,Obtaining),
            (Stolen_Infor,From_User),
            (Coersive_Force,From_User),
            (Cross_Site,From_User),
            (Key_Logger,From_User),
            (Social,From_User),
            (Obtaining,Authen_Vio)
           )
```

```python
net = Network(G)
```

```python
# net.help_generate_NPT(G.list_nodes, 'data/demo')
```

```python
for node in G.list_nodes:
    if node.isFunc == False:
        net.set_NPT_from_csv(node, f'data/demo/{node.id_name}.csv')
    else:
        net.set_NPT_func(node)
```

### Small Net

```python
Encrypted_File = Boolean_Node('Encrypted_File')
File_Decrypted = Boolean_Node('File_Decrypted')
PlainText_File = Boolean_Node('PlainText_File')

Stolen_Infor = Boolean_Node('Stolen_Infor')
Cross_Site = Boolean_Node('Cross_Site')
Social = Boolean_Node('Social')

Credential_Brute = Boolean_Node('Credential_Brute')

Obtaining = OR_Node('Obtaining')
From_User = OR_Node('From_User')
From_System = XOR_Node('From_System')
```

```python
G = Graph()
G.add_nodes(Encrypted_File,File_Decrypted,PlainText_File,From_System,Credential_Brute,From_User,Stolen_Infor,Cross_Site,Social,Obtaining)
```

```python
G.add_edges((Encrypted_File,File_Decrypted),
            (PlainText_File,From_System),
            (File_Decrypted,From_System),
            (From_System,Obtaining),
            
            (Credential_Brute,Obtaining),
            (From_User,Obtaining),
            
            (Stolen_Infor,From_User),
            (Cross_Site,From_User),
            (Social,From_User),
           )
```

```python
G.visualize(figsize = (16,8),node_size = 6000)
```

```python
net = Network(G)
```

```python
for node in G.list_nodes:
    if node.isFunc == False:
        net.set_NPT_from_csv(node, f'data/demo/{node.id_name}.csv')
    else:
        net.set_NPT_func(node)
```
